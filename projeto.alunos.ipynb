```python
# Tratando os dados NaN do DataFrame.
dados.isnull()
dados = dados.fillna(0) 
dados
```

```python
# Retirando Carlos e Alice da tabela, pois n칚o fazem mais parte da classe.
dados.drop([8, 7], axis=0, inplace=True)
dados
```

```python
# Selecionando somente os alunos aprovados.
alunos_aprovados = dados['Aprovado'] == True
```

```python
alunos_aprovados = dados[alunos_aprovados]
```

```python
alunos_aprovados
```

```python
# Trocando as notas 7.0 da tabela pelo o 8.0.
alunos_aprovados.replace(7.0, 8.0)
```

```python
alunos_aprovados = alunos_aprovados.replace(7.0, 8.0)
```

```python
# Salvando o arquivo em .csv.
alunos_aprovados.to_csv("Alunos_aprovados.csv", index=False)
```

```python
# Adicionando ponto extra de 40% da nota para todos os alunos.
dados["ponto_extra"] = dados["Notas"].apply(lambda x: x * 0.4)
dados
```

```python
# Somando as notas com os pontos extras para obter as notas finais.
dados["Notas_finais"] = dados["Notas"] + dados["ponto_extra"]
dados
```

```python
# Fun칞칚o lambda para exibir True caso a nota final seja >= 6 e Falso caso contr치rio.
dados["Aprovado_final"] = dados["Notas_finais"].apply(lambda x: True if x >= 6 else False)
dados
```

```python
selecao = (dados['Aprovado'] == False) & (dados['Aprovado_final'] == True)
dados[selecao]
```
